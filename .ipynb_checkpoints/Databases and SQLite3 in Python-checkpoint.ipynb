## Example SQL Query

```python
US_sold_query = """
WITH usa_sold AS
    (
    SELECT 
        il.invoice_id, 
        il.track_id, 
        i.billing_country, 
        il.unit_price*il.quantity total_sold,
        t.genre_id
    FROM invoice_line il
    INNER JOIN invoice i ON il.invoice_id = i.invoice_id
    INNER JOIN track t ON il.track_id = t.track_id
    WHERE i.billing_country = 'USA'
    )
 
SELECT 
    u.billing_country, 
    g.name, 
    CAST(SUM(total_sold) AS INT) dollars_sold
FROM usa_sold u
INNER JOIN genre g ON u.genre_id = g.genre_id
GROUP BY g.name
ORDER BY dollars_sold DESC;"""
```

### Transforming the data into a dataframe

```python
US_sold_data = conn.execute(US_sold_query).fetchall()
US_sold_df = pd.DataFrame(US_sold_data, columns =['Country','Name','Dollars Sold'])
def change_country(row):
    if row["Dollars Sold"] < 20:
        row["Name"] = "Other"
    return row;
US_sold_df = US_sold_df.apply(change_country, axis = 1)
US_sold_df = US_sold_df.groupby("Name").sum().sort_values("Dollars Sold", ascending = False).reset_index()
#figure out groupby and differences in data manipulation between SQL and python
```

### Some basic plotting with Seaborn

```python
fig, ax = plt.subplots(figsize = (15,5))
#can also do plt.figure(figsize = (15,5))
sns.barplot(x = "Name", y = "Dollars Sold", data = US_sold_df)
sns.set(style="darkgrid")
```

```python
%%sql

WITH agent_cust AS
(
SELECT 
    (e.first_name || " " || e.last_name) employee,
    e.title,
    c.customer_id
FROM employee e
INNER JOIN customer c ON e.employee_id = c.support_rep_id
WHERE e.title = 'Sales Support Agent'
)

SELECT 
    a.employee,
    a.title,
    SUM(i.total) total_sales
FROM invoice i
INNER JOIN agent_cust a ON i.customer_id = a.customer_id
GROUP BY a.employee;
```

```python
WITH test AS
(
SELECT 
    COUNT(DISTINCT c.customer_id) num_of_customers,
    c.country,
    COUNT(i.invoice_id) num_of_orders,
    ROUND(SUM(i.total)) total_sales
FROM customer c
INNER JOIN invoice i ON c.customer_id = i.customer_id
GROUP BY c.country
ORDER BY total_sales DESC
)

SELECT  
    t.country,
    CASE 
        WHEN num_of_customers = 1 
        THEN "Other" 
        ELSE num_of_customers 
        END AS num_of_customers_filt,
    t.num_of_orders,
    total_sales,
    ROUND(total_sales/num_of_orders, 2) avg_sale
    GROUP BY
FROM test t
```
